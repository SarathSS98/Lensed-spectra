```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import pylab as pl
```

```python
df=pd.read_csv('whtfile.csv',header=0)
```

```python
for i in range(len(df)):
    x=['1219','1549','1900','2800']
    y1 = [df.Ratio_LyA_High[i],df.Ratio_CIV_High[i],df.Ratio_CIII_High[i],df.Ratio_MgII_High[i]]
    y2 = [df.Ratio_LyA_Conti[i],df.Ratio_CIV_Conti[i],df.Ratio_CIII_Conti[i],df.Ratio_MgII_Conti[i]]
    plt.plot(x,y1,'bo',ls='--',label='line')
    plt.plot(x,y2,'ko',ls='--',label='continuum')
    plt.legend()
    text=[df.name[i]]
  
  #  plt.ylim(0.1,10)
    plt.title('Ratio of Continnum and Emission Lines')
    plt.xlabel('Wavelength')
    plt.ylabel('Ratios')
    plt.savefig('/mnt/data/sarath/UNAB-PHD/Semester-1/Project-lensed Quasars/new lenses/ratio plots/'+df.name[i])
    plt.show()
```
